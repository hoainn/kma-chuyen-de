# Anomaly Detection: Isolation Forest vs Variational Autoencoder
### DongTing Dataset — Linux Kernel Syscall Anomaly Detection

---

Adapted from the reference notebook (credit-card fraud) to the **DongTing** dataset.

> G. Duan et al., *DongTing: A large-scale dataset for anomaly detection of the Linux kernel*, JSS 2023.

**Dataset stats:**  
- 18,966 labeled syscall sequences (6,850 normal + 12,116 attack)  
- Pre-split 80/10/10 (train/val/test) by the dataset authors  
- Attack sequences trigger 26 different Linux kernel bug types

**Goal:** detect attack (anomalous) syscall sequences using **unsupervised** methods,
trained only on normal sequences.

**Methods compared:**
1. Isolation Forest (sklearn)
2. Variational Autoencoder (TensorFlow/Keras)

---
## 1. Setup & Imports

```python
import os, warnings, json, re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
warnings.filterwarnings('ignore')

from pymongo import MongoClient
from tqdm import tqdm

from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.metrics import (
    classification_report, confusion_matrix,
    roc_auc_score, average_precision_score,
    roc_curve, precision_recall_curve, f1_score
)
from sklearn.model_selection import train_test_split

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, Model
from tensorflow.keras.losses import MeanSquaredError

np.random.seed(42)
tf.random.set_seed(42)

MONGO_URI = os.environ.get('MONGO_URI', 'mongodb://mongo:27017/')
MONGO_DB  = os.environ.get('MONGO_DB',  'syzbot_DB')

print(f'TensorFlow: {tf.__version__}')
print(f'MongoDB URI: {MONGO_URI}')
```

---
## 2. Load DongTing from MongoDB

### Collections
| Collection | Content |
|---|---|
| `kernel_convert_baseline` | Metadata + train/val/test split for all 18,966 sequences |
| `kernel_syscall_normal_strace` | Normal sequences — field `kns_normal_mlseq_list` (pipe-sep IDs) |
| `kernel_syscallhook_bugpoc_trace_sum` | Attack sequences — field `kshs_bugpoc_syscall_mlcode` (pipe-sep IDs) |

Feature engineering: each sequence → **relative frequency vector** over the top-*N* syscall IDs.

```python
client = MongoClient(MONGO_URI)
db = client[MONGO_DB]

col_baseline = db['kernel_convert_baseline']
col_normal   = db['kernel_syscall_normal_strace']
col_attack   = db['kernel_syscallhook_bugpoc_trace_sum']

# Check what's imported
counts = {
    'baseline':  col_baseline.estimated_document_count(),
    'normal_seq': col_normal.estimated_document_count(),
    'attack_seq': col_attack.estimated_document_count(),
}
print('Collection sizes:', counts)
HAS_SEQUENCES = counts['normal_seq'] > 0 and counts['attack_seq'] > 0
print(f'Full sequence data available: {HAS_SEQUENCES}')
```

```python
# ── Load baseline metadata ────────────────────────────────────────────────────
print('Loading baseline metadata...')
baseline_docs = list(col_baseline.find({}, {'_id': 0}))
df_meta = pd.DataFrame(baseline_docs)

# Normalize $numberInt fields from bsondump encoding
def unwrap_int(v):
    if isinstance(v, dict) and '$numberInt' in v:
        return int(v['$numberInt'])
    return v

df_meta['syscall_counts'] = df_meta['kcb_syscall_counts'].apply(unwrap_int).fillna(0).astype(int)
df_meta['syscall_sizes']  = df_meta['kcb_syscall_sizes'].apply(unwrap_int).fillna(0).astype(int)
df_meta['label']          = (df_meta['kcb_seq_lables'] == 'Attach').astype(int)
df_meta['split']          = df_meta['kcb_seq_class'].map({
    'DTDS-train': 'train', 'DTDS-validation': 'val', 'DTDS-test': 'test'
})

# One-hot kernel version
ver_dummies = pd.get_dummies(df_meta['kcb_master_line_ver'], prefix='ver')
df_meta = pd.concat([df_meta, ver_dummies], axis=1)

print(df_meta[['kcb_seq_id','label','split','syscall_counts','syscall_sizes']].head(3))
print(f'\nShape: {df_meta.shape}')
print(df_meta['label'].value_counts().rename({0:'Normal',1:'Attack'}))
```

```python
# ── Build frequency-vector features from sequence collections ─────────────────
# Parse a pipe-separated ML-code string into a list of integers
def parse_mlcode(s):
    if not s or not isinstance(s, str) or s.startswith('sy_') or s.startswith('ml_'):
        return []
    return [int(x) for x in s.split('|') if x.strip().lstrip('-').isdigit()]

# Gather all syscall IDs to find the top-N most common
SAMPLE_LIMIT = 5000  # process at most N sequences per class for speed

all_ids = []
if HAS_SEQUENCES:
    print('Scanning syscall ID distribution (sample)...')
    for doc in col_normal.find({}, {'kns_normal_mlseq_list': 1}, limit=SAMPLE_LIMIT):
        all_ids.extend(parse_mlcode(doc.get('kns_normal_mlseq_list', '')))
    for doc in col_attack.find({}, {'kshs_bugpoc_syscall_mlcode': 1}, limit=SAMPLE_LIMIT):
        all_ids.extend(parse_mlcode(doc.get('kshs_bugpoc_syscall_mlcode', '')))

    id_counts = pd.Series(all_ids).value_counts()
    TOP_N = 50
    top_ids = id_counts.head(TOP_N).index.tolist()
    print(f'Top-{TOP_N} syscall IDs: {top_ids[:10]} ...')
else:
    top_ids = []
    print('No sequence collections — will use metadata-only features.')
```

```python
# ── Build feature matrix ──────────────────────────────────────────────────────
def make_freq_vector(mlcode_str, top_ids):
    ids = parse_mlcode(mlcode_str)
    if not ids:
        return np.zeros(len(top_ids))
    total = len(ids)
    return np.array([ids.count(sid) / total for sid in top_ids], dtype=np.float32)

def build_features(df_meta, col_normal, col_attack, top_ids):
    ver_cols = [c for c in df_meta.columns if c.startswith('ver_')]
    rows = []

    if HAS_SEQUENCES:
        # Build lookup dictionaries from sequence collections
        print('Building sequence lookup tables...')
        normal_lookup = {
            doc['kns_normal_file_name']: doc.get('kns_normal_mlseq_list', '')
            for doc in tqdm(col_normal.find({}, {'kns_normal_file_name': 1, 'kns_normal_mlseq_list': 1}),
                            desc='normal', total=col_normal.estimated_document_count())
        }
        attack_lookup = {
            doc['kshs_poclog_name']: doc.get('kshs_bugpoc_syscall_mlcode', '')
            for doc in tqdm(col_attack.find({}, {'kshs_poclog_name': 1, 'kshs_bugpoc_syscall_mlcode': 1}),
                            desc='attack', total=col_attack.estimated_document_count())
        }

    for _, row in tqdm(df_meta.iterrows(), total=len(df_meta), desc='features'):
        ver_feats = row[ver_cols].values.astype(np.float32)
        cnt = np.log1p(row['syscall_counts'])
        sz  = np.log1p(row['syscall_sizes'])
        meta_feats = np.array([cnt, sz], dtype=np.float32)

        if HAS_SEQUENCES:
            bug_name = row['kcb_bug_name']
            if row['label'] == 0:  # Normal
                fname = bug_name.rsplit('.', 1)[0]  # strip .log
                mlcode = normal_lookup.get(fname, '')
            else:  # Attack
                mlcode = attack_lookup.get(bug_name, '')
            freq_feats = make_freq_vector(mlcode, top_ids)
            feat = np.concatenate([freq_feats, meta_feats, ver_feats])
        else:
            feat = np.concatenate([meta_feats, ver_feats])

        rows.append(feat)

    return np.array(rows, dtype=np.float32)

print('Building feature matrix...')
X_all = build_features(df_meta, col_normal, col_attack, top_ids)
y_all = df_meta['label'].values
split_all = df_meta['split'].values

print(f'Feature matrix shape: {X_all.shape}')
print(f'Labels — Normal: {(y_all==0).sum()}, Attack: {(y_all==1).sum()}')
```

## 4. Preprocessing

1. Use the dataset's own 80/10/10 train/val/test split
2. Scale all numeric features with `StandardScaler` fit on **training normal sequences only**
3. Model trains on normal sequences; attack sequences are seen only at test time

```python
train_mask = split_all == 'train'
val_mask   = split_all == 'val'
test_mask  = split_all == 'test'

X_train_full, y_train_full = X_all[train_mask], y_all[train_mask]
X_val,        y_val        = X_all[val_mask],   y_all[val_mask]
X_test,       y_test       = X_all[test_mask],  y_all[test_mask]

# Fit scaler only on normal training sequences
scaler = StandardScaler()
X_train_normal = X_train_full[y_train_full == 0]
scaler.fit(X_train_normal)

X_train = scaler.transform(X_train_full)
X_val   = scaler.transform(X_val)
X_test  = scaler.transform(X_test)

# Normal-only training set (for VAE unsupervised training)
X_train_vae = X_train[y_train_full == 0]

print(f'Train: {X_train.shape} | Val: {X_val.shape} | Test: {X_test.shape}')
print(f'Train normal (VAE input): {X_train_vae.shape}')
print(f'Test — Normal: {(y_test==0).sum()}, Attack: {(y_test==1).sum()}')
```

## 5. Isolation Forest

### How It Works

Isolation Forest builds an ensemble of random decision trees that recursively
partition the feature space. **Anomalous points** (attacks) are isolated in fewer
splits — they land in shorter paths — giving higher anomaly scores.

**Key parameter:** `contamination` = expected fraction of anomalies.  
We use the true attack rate in the training set.

```python
fraud_rate = y_train_full.mean()  # true attack fraction in training set
print(f'Attack rate in training set: {fraud_rate:.4f}')

iso = IsolationForest(
    n_estimators=200,
    contamination=float(fraud_rate),
    max_samples='auto',
    random_state=42,
    n_jobs=-1,
)
iso.fit(X_train)
print('Isolation Forest trained.')
```

```python
# Anomaly scores: lower decision_function → more anomalous → negate for consistent direction
iso_scores = -iso.decision_function(X_test)          # higher = more anomalous
iso_preds  = (iso.predict(X_test) == -1).astype(int) # -1 = anomaly

print('=== Isolation Forest Classification Report ===')
print(classification_report(y_test, iso_preds, target_names=['Normal', 'Attack']))
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# 1. Confusion matrix
cm = confusion_matrix(y_test, iso_preds)
sns.heatmap(cm, annot=True, fmt='d', ax=axes[0],
            xticklabels=['Normal', 'Attack'], yticklabels=['Normal', 'Attack'],
            cmap='Blues', linewidths=0.5)
axes[0].set_title('Isolation Forest\nConfusion Matrix')
axes[0].set_ylabel('True'); axes[0].set_xlabel('Predicted')

# 2. ROC curve
fpr, tpr, _ = roc_curve(y_test, iso_scores)
auc_iso = roc_auc_score(y_test, iso_scores)
axes[1].plot(fpr, tpr, color='#4C72B0', label=f'AUC = {auc_iso:.3f}')
axes[1].plot([0,1],[0,1],'k--', alpha=0.4)
axes[1].set_title('Isolation Forest — ROC Curve')
axes[1].set_xlabel('False Positive Rate'); axes[1].set_ylabel('True Positive Rate')
axes[1].legend()

# 3. Precision-Recall curve
prec, rec, _ = precision_recall_curve(y_test, iso_scores)
ap_iso = average_precision_score(y_test, iso_scores)
axes[2].plot(rec, prec, color='#DD8452', label=f'AP = {ap_iso:.3f}')
axes[2].set_title('Isolation Forest — PR Curve')
axes[2].set_xlabel('Recall'); axes[2].set_ylabel('Precision')
axes[2].legend()

plt.suptitle('Isolation Forest Results', fontsize=13)
plt.tight_layout()
plt.savefig('outputs/iso_results.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Anomaly score distribution
fig, ax = plt.subplots(figsize=(12, 4))
iso_normal_scores = iso_scores[y_test == 0]
iso_attack_scores = iso_scores[y_test == 1]
ax.hist(iso_normal_scores, bins=60, color='#4C72B0', alpha=0.6, label='Normal', density=True)
ax.hist(iso_attack_scores, bins=60, color='#DD8452', alpha=0.6, label='Attack', density=True)
ax.set_title('Isolation Forest — Anomaly Score Distribution')
ax.set_xlabel('Anomaly Score (higher = more anomalous)')
ax.set_ylabel('Density')
ax.legend()
plt.tight_layout()
plt.savefig('outputs/iso_score_dist.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# PCA visualization of Isolation Forest decisions
pca = PCA(n_components=2, random_state=42)
X_test_2d = pca.fit_transform(X_test)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
for ax, labels_to_plot, title in [
    (axes[0], y_test,       'True Labels'),
    (axes[1], iso_preds,    'IF Predictions'),
]:
    sc = ax.scatter(X_test_2d[:, 0], X_test_2d[:, 1],
                    c=labels_to_plot, cmap='coolwarm',
                    s=5, alpha=0.4)
    ax.set_title(title); ax.set_xlabel('PC1'); ax.set_ylabel('PC2')
    plt.colorbar(sc, ax=ax, label='0=Normal, 1=Attack')

plt.suptitle('PCA Visualization — Isolation Forest (test set)', fontsize=13)
plt.tight_layout()
plt.savefig('outputs/iso_pca.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 6. Variational Autoencoder (VAE)

### How It Works

A VAE encodes normal sequences into a compressed **latent space** and learns to
reconstruct them. At inference, attack sequences reconstruct poorly because the
model has never seen them — their **reconstruction error** is high.

Architecture (from DeSFAM paper, Section IV-D):
- Encoder: Input → Hidden → μ, σ (latent)
- Decoder: Latent z → Hidden → Reconstruction
- Activations: SELU, Dropout = 0.2, L2 weight decay
- Loss: MSE reconstruction + β·KL-divergence

```python
INPUT_DIM  = X_train.shape[1]
HIDDEN_DIM = 32
LATENT_DIM = 16
BETA       = 1.0

# ── Encoder ──────────────────────────────────────────────────────────────────
enc_input = keras.Input(shape=(INPUT_DIM,), name='encoder_input')
x = layers.Dense(HIDDEN_DIM, activation='selu',
                 kernel_regularizer=keras.regularizers.l2(1e-4))(enc_input)
x = layers.Dropout(0.2)(x)
z_mean    = layers.Dense(LATENT_DIM, name='z_mean')(x)
z_log_var = layers.Dense(LATENT_DIM, name='z_log_var')(x)

# Reparameterization trick
def sampling(args):
    z_mean, z_log_var = args
    eps = tf.random.normal(tf.shape(z_mean))
    return z_mean + tf.exp(0.5 * z_log_var) * eps

z = layers.Lambda(sampling, name='z')([z_mean, z_log_var])
encoder = Model(enc_input, [z_mean, z_log_var, z], name='encoder')

# ── Decoder ──────────────────────────────────────────────────────────────────
dec_input = keras.Input(shape=(LATENT_DIM,), name='decoder_input')
x = layers.Dense(HIDDEN_DIM, activation='selu',
                 kernel_regularizer=keras.regularizers.l2(1e-4))(dec_input)
x = layers.Dropout(0.2)(x)
dec_output = layers.Dense(INPUT_DIM)(x)
decoder = Model(dec_input, dec_output, name='decoder')

encoder.summary()
decoder.summary()
```

```python
# ── VAE model with custom training step ──────────────────────────────────────
class VAE(Model):
    def __init__(self, encoder, decoder, beta=1.0, **kwargs):
        super().__init__(**kwargs)
        self.encoder = encoder
        self.decoder = decoder
        self.beta = beta
        self.mse_loss_tracker   = keras.metrics.Mean(name='mse_loss')
        self.kl_loss_tracker    = keras.metrics.Mean(name='kl_loss')
        self.total_loss_tracker = keras.metrics.Mean(name='total_loss')

    @property
    def metrics(self):
        return [self.total_loss_tracker, self.mse_loss_tracker, self.kl_loss_tracker]

    def train_step(self, data):
        if isinstance(data, tuple):
            data = data[0]
        with tf.GradientTape() as tape:
            z_mean, z_log_var, z = self.encoder(data, training=True)
            reconstruction = self.decoder(z, training=True)
            mse_loss = tf.reduce_mean(tf.reduce_sum(
                MeanSquaredError(reduction='none')(data, reconstruction), axis=-1
            ))
            kl_loss = -0.5 * tf.reduce_mean(
                1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var)
            )
            total = mse_loss + self.beta * kl_loss
        grads = tape.gradient(total, self.trainable_weights)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))
        self.mse_loss_tracker.update_state(mse_loss)
        self.kl_loss_tracker.update_state(kl_loss)
        self.total_loss_tracker.update_state(total)
        return {m.name: m.result() for m in self.metrics}

vae = VAE(encoder, decoder, beta=BETA)
vae.compile(optimizer=keras.optimizers.Adam(learning_rate=1e-3))
print('VAE compiled.')
```

```python
# Manual validation split — 10% of normal training set
val_size = max(1, int(len(X_train_vae) * 0.1))
X_val_vae   = X_train_vae[:val_size]
X_train_vae_fit = X_train_vae[val_size:]

early_stop = keras.callbacks.EarlyStopping(
    monitor='val_loss', patience=10, restore_best_weights=True
)
lr_sched = keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss', factor=0.5, patience=5, verbose=0
)

history = vae.fit(
    X_train_vae_fit, X_train_vae_fit,
    epochs=80,
    batch_size=256,
    validation_data=(X_val_vae, X_val_vae),
    callbacks=[early_stop, lr_sched],
    verbose=1,
)
print('VAE training complete.')
```

```python
# Training curves
fig, axes = plt.subplots(1, 3, figsize=(16, 4))
for ax, key, title in [
    (axes[0], 'total_loss', 'Total Loss'),
    (axes[1], 'mse_loss',   'MSE Reconstruction Loss'),
    (axes[2], 'kl_loss',    'KL Divergence'),
]:
    if key in history.history:
        ax.plot(history.history[key], label='Train')
        val_key = f'val_{key}'
        if val_key in history.history:
            ax.plot(history.history[val_key], label='Val')
    ax.set_title(title)
    ax.set_xlabel('Epoch')
    ax.legend()

plt.suptitle('VAE Training History', fontsize=13)
plt.tight_layout()
plt.savefig('outputs/vae_training.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ── VAE anomaly scoring ───────────────────────────────────────────────────────
# Reconstruction error per sample (sum of squared errors across features)
def vae_reconstruction_error(X):
    z_mean_p, z_log_var_p, z_p = encoder.predict(X, verbose=0)
    X_recon = decoder.predict(z_p, verbose=0)
    return np.mean((X - X_recon) ** 2, axis=1)  # mean squared error per sample

# Threshold = 99.5th percentile of training normal errors (from DeSFAM paper)
train_errors = vae_reconstruction_error(X_train_vae)
threshold    = np.percentile(train_errors, 99.5)
print(f'VAE threshold (99.5th pct of normal training errors): {threshold:.6f}')

vae_scores = vae_reconstruction_error(X_test)
vae_preds  = (vae_scores > threshold).astype(int)

print('=== VAE Classification Report ===')
print(classification_report(y_test, vae_preds, target_names=['Normal', 'Attack']))
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Confusion matrix
cm_vae = confusion_matrix(y_test, vae_preds)
sns.heatmap(cm_vae, annot=True, fmt='d', ax=axes[0],
            xticklabels=['Normal', 'Attack'], yticklabels=['Normal', 'Attack'],
            cmap='Oranges', linewidths=0.5)
axes[0].set_title('VAE — Confusion Matrix')
axes[0].set_ylabel('True'); axes[0].set_xlabel('Predicted')

# ROC
fpr_v, tpr_v, _ = roc_curve(y_test, vae_scores)
auc_vae = roc_auc_score(y_test, vae_scores)
axes[1].plot(fpr_v, tpr_v, color='#DD8452', label=f'AUC = {auc_vae:.3f}')
axes[1].plot([0,1],[0,1],'k--', alpha=0.4)
axes[1].set_title('VAE — ROC Curve')
axes[1].set_xlabel('False Positive Rate'); axes[1].set_ylabel('True Positive Rate')
axes[1].legend()

# PR
prec_v, rec_v, _ = precision_recall_curve(y_test, vae_scores)
ap_vae = average_precision_score(y_test, vae_scores)
axes[2].plot(rec_v, prec_v, color='#4C72B0', label=f'AP = {ap_vae:.3f}')
axes[2].set_title('VAE — Precision-Recall Curve')
axes[2].set_xlabel('Recall'); axes[2].set_ylabel('Precision')
axes[2].legend()

plt.suptitle('VAE Results', fontsize=13)
plt.tight_layout()
plt.savefig('outputs/vae_results.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Latent space visualization
z_mean_test, _, _ = encoder.predict(X_test, verbose=0)
z_2d = PCA(n_components=2, random_state=42).fit_transform(z_mean_test)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
for ax, labels_to_plot, title in [
    (axes[0], y_test,    'True Labels'),
    (axes[1], vae_preds, 'VAE Predictions'),
]:
    sc = ax.scatter(z_2d[:, 0], z_2d[:, 1],
                    c=labels_to_plot, cmap='coolwarm',
                    s=5, alpha=0.4)
    ax.set_title(f'VAE Latent Space (PCA) — {title}')
    ax.set_xlabel('PC1'); ax.set_ylabel('PC2')
    plt.colorbar(sc, ax=ax, label='0=Normal, 1=Attack')

plt.tight_layout()
plt.savefig('outputs/vae_latent.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 7. Side-by-Side Comparison

| Aspect | Isolation Forest | VAE |
|---|---|---|
| Training speed | Very fast | Slower (neural network) |
| Interpretability | Low | Low (latent space) |
| Handles high-dim | Moderate | Good |
| Zero-day detection | Good | Good (reconstructed latent) |
| Requires normal only | Yes | Yes |

```python
from sklearn.metrics import precision_score, recall_score, accuracy_score

results = pd.DataFrame({
    'Model':     ['Isolation Forest', 'VAE'],
    'ROC-AUC':   [auc_iso, auc_vae],
    'Avg Prec':  [ap_iso,  ap_vae],
    'F1 Score':  [
        f1_score(y_test, iso_preds, zero_division=0),
        f1_score(y_test, vae_preds, zero_division=0),
    ],
    'Precision': [
        precision_score(y_test, iso_preds, zero_division=0),
        precision_score(y_test, vae_preds, zero_division=0),
    ],
    'Recall': [
        recall_score(y_test, iso_preds, zero_division=0),
        recall_score(y_test, vae_preds, zero_division=0),
    ],
    'Accuracy': [
        accuracy_score(y_test, iso_preds),
        accuracy_score(y_test, vae_preds),
    ],
}).set_index('Model')

print(results.round(4).to_string())
```

```python
# ROC & PR side-by-side
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# ROC
axes[0].plot(fpr,   tpr,   color='#4C72B0', label=f'IF  (AUC={auc_iso:.3f})')
axes[0].plot(fpr_v, tpr_v, color='#DD8452', label=f'VAE (AUC={auc_vae:.3f})')
axes[0].plot([0,1],[0,1],'k--', alpha=0.4)
axes[0].set_title('ROC Curves — IF vs VAE'); axes[0].legend()
axes[0].set_xlabel('False Positive Rate'); axes[0].set_ylabel('True Positive Rate')

# PR
axes[1].plot(rec,   prec,   color='#4C72B0', label=f'IF  (AP={ap_iso:.3f})')
axes[1].plot(rec_v, prec_v, color='#DD8452', label=f'VAE (AP={ap_vae:.3f})')
axes[1].set_title('Precision-Recall Curves — IF vs VAE'); axes[1].legend()
axes[1].set_xlabel('Recall'); axes[1].set_ylabel('Precision')

plt.suptitle('IF vs VAE — DongTing Dataset', fontsize=13)
plt.tight_layout()
plt.savefig('outputs/comparison_roc_pr.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Normalized anomaly score overlay
from sklearn.preprocessing import MinMaxScaler as MMS
mm = MMS()
iso_norm = mm.fit_transform(iso_scores.reshape(-1,1)).ravel()
mm2 = MMS()
vae_norm = mm2.fit_transform(vae_scores.reshape(-1,1)).ravel()

fig, axes = plt.subplots(1, 2, figsize=(14, 4))
for ax, norm_scores, title in [
    (axes[0], iso_norm, 'Isolation Forest'),
    (axes[1], vae_norm, 'VAE'),
]:
    ax.hist(norm_scores[y_test==0], bins=50, color='#4C72B0', alpha=0.6, label='Normal', density=True)
    ax.hist(norm_scores[y_test==1], bins=50, color='#DD8452', alpha=0.6, label='Attack', density=True)
    ax.set_title(f'{title} — Normalized Score Distribution')
    ax.set_xlabel('Normalized Anomaly Score')
    ax.legend()

plt.tight_layout()
plt.savefig('outputs/comparison_score_dist.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ── Ensemble: weighted combination (α=0.7 VAE, per DeSFAM paper) ─────────────
ALPHA = 0.7
ensemble_scores = ALPHA * vae_norm + (1 - ALPHA) * iso_norm

# Threshold at 99.5th percentile of normal scores
ens_threshold = np.percentile(ensemble_scores[y_test == 0], 99.5)
ensemble_preds = (ensemble_scores > ens_threshold).astype(int)

ens_auc = roc_auc_score(y_test, ensemble_scores)
ens_ap  = average_precision_score(y_test, ensemble_scores)
ens_f1  = f1_score(y_test, ensemble_preds, zero_division=0)

print(f'Ensemble (α={ALPHA}) — AUC: {ens_auc:.4f} | AP: {ens_ap:.4f} | F1: {ens_f1:.4f}')
print(classification_report(y_test, ensemble_preds, target_names=['Normal', 'Attack']))
```

```python
# Final summary
all_models = {
    'Isolation Forest': (iso_scores,      iso_preds),
    'VAE':              (vae_scores,      vae_preds),
    f'Ensemble α={ALPHA}': (ensemble_scores, ensemble_preds),
}

summary_rows = []
for name, (scores, preds) in all_models.items():
    summary_rows.append({
        'Model':     name,
        'ROC-AUC':   round(roc_auc_score(y_test, scores), 4),
        'Avg Prec':  round(average_precision_score(y_test, scores), 4),
        'F1':        round(f1_score(y_test, preds, zero_division=0), 4),
        'Precision': round(precision_score(y_test, preds, zero_division=0), 4),
        'Recall':    round(recall_score(y_test, preds, zero_division=0), 4),
        'Accuracy':  round(accuracy_score(y_test, preds), 4),
    })

df_summary = pd.DataFrame(summary_rows).set_index('Model')
print('\n=== Final Summary ===')
print(df_summary.to_string())

df_summary.to_csv('outputs/summary.csv')
print('\nResults saved to outputs/summary.csv')
```

## 8. Key Takeaways

### DongTing vs Credit Card Fraud

| Aspect | Credit Card (reference) | DongTing (this notebook) |
|---|---|---|
| Feature space | 30 (PCA-financial) | Syscall freq + metadata |
| Anomaly rate | ~0.17% | ~64% |
| Sequence length | Single transaction | 61–100K+ syscalls |
| Context | Financial fraud | Kernel exploits |

### When to Use Each Model

| Aspect | Isolation Forest | VAE |
|---|---|---|
| Training speed | Very fast | Slower |
| No labeled data needed | Yes | Yes |
| Captures sequence structure | No | Partially |
| Scales to large data | Yes | Yes |
| Best used when | Quick baseline, tabular features | Complex nonlinear patterns |

### DeSFAM Approach
The DeSFAM paper (Phase 2) combines both: **α·A_VAE + (1-α)·A_IF** with α=0.7,
reporting 94% precision and 90% recall on DongTing sequences captured live via eBPF.
